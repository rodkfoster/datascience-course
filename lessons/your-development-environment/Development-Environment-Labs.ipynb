### Common Commands
<a id='commands'></a>


Lets start by overviewing some common commands to use today. For a more extensive list please reference [here](https://www.cheatography.com/davechild/cheat-sheets/linux-command-line/) as one of many possible resources.

#### General format for commands

`<command> -<options> <arguments>`
* `<command>` is the action we want the computer to take
* `<options>` (or "flags") modify the behavior of the command
* `<arguments>` are the things we want the command to act on

For Linux and Mac users, you can get view the **man**ual for a command by typing `man <command>`. For Windows users, you can view the help page by typing `<command> --help`.

#### Commands used in this lesson 
- **cd** : Change directory
    - `cd`     Alone returns you to the home directory
    - `cd <path>` **c**hanges **d**irectory to the path you specify, which can be a relative path or an absolute path
    - `cd ~`   References the home directory as a starting point
    - `cd .`   References the current directory as a starting point
    - `cd ..`  Asks the computer to move *up* one directory
    - `cd -`   Takes you back to the previous location
    
    
- **pwd** : Print Working Directory - this commmand shows you where you are by returning the current path to the directory you are in

- **clear** clears all output from your console

- **ls** : Lists the contents of the current directory
    - `ls -l`  -l (-l is character not number one) is a flag asking the computer to show file or directory, size, modified date and time, file or folder name and owner of file and its permission
    - `ls -a`  -a List all files including hidden file starting with ‘.‘.
    - `ls lh`  -With combination of -lh the return shows sizes in human readable format.
    - `ls *` also lists the contents of subdirectories (one level deep) in your working directory)
    - `ls <path>` lists files in a specific directory (without changing your working directory)
    - Use `ls -help` (PC) or `man ls` (apple) to figure out what these arguments do
        - `ls -F` 
        - `ls -r`
        
        
- **mkdir** : Make Directory
    - `mkdir NewFolder` will create a folder in the current directory.
    - To create a folder in a specific directory path, specify the directory, such as `mkdir /c/Documents/NewFolder`
    - To create a parent folder to house a new directory, use `mkdir –p /c/NewParentFolder/NewFolder`


- **touch** : creates a new file.
    - `touch myFile.txt` will create a text file in the new directory.
    - `touch myFile.md` will create a markdown file in the current directory.
    - Multiple files can be created at the same time. `touch Super.txt Califragilistic.txt Expialidocious.txt`
    

- **rm** : removes a file (**Note**: Be careful as once deleted your information is gone. Practice in a sub-directory you do not care about)
    - `rmdir FolderName` will remove an empty folder.

    - `rm –rf –FolderName` will remove the folder and all contents.

    - `rm FileName` will remove the file in the current directory.

    - `rm –i FileName` will prompt you for confirmation before removing the file.
    - `rm -ir <dirname>` removes a directory and **r**ecursively deletes all of its contents

- **mv** : moves files to a new location or renames a file
    - `mv <filename> <new path>` **m**o**v**es a file from its current location to `<new path>`
    - `mv <filename> <new filename>` renames a file without changing its location
    
- **cp** Copies a file
    - `cp <filename> <new path>` **c**o**p**ies a file from its current location to `<new path>`, leaving the original file unchanged
    - `cp <filename> <new filename>` copies a file without changing its location



    
    
- Viewing a file - Sometimes it's nice to view the contents of files as text. There are a variety of ways to do this. The commands **cat**, **head**, and **tail** will allow us to view the entire or partial contents of a target file.
    - `cat myFile.txt` will display short files.

    - `head myFile.txt` will show the first few lines.

    - `head -n myFile.txt` will show the first n lines.

    - `tail myFile.txt` will show the last few lines.

    - `tail –n myFile.txt` will show the last n lines
    
    - `echo myFile.txt` 


#### Intermediate commands

- **`less`** : Returns less text
     - `less <filename>` allows you to page or scroll through the file
    - Hit the spacebar to go down a page, use the arrow keys to scroll up and down, and hit `q` to exit.

- **`wc`** : Counts words
     - `wc <filename>` returns the **c**ount of lines, **w**ords, and characters in a file
    - `wc -l <filename>` only counts lines, `wc -w <filename>` only counts words, and `wc -c <filename>` only counts characters
    - A "word" is defined as any set of characters delimited by a space.

- **`find`** Finds a file and returns it's path
     - `find <path> -name <name>` will recursively search the specified path (and its subdirectories) and **find** files and directories with a given `<name>`
        - Use `.` for the `<path>` to refer to the working directory.
    - For the `<name>`, you can search for an exact match, or use wildcard characters to search for a partial match:
        - `*` specifies any number of any characters, such as `find . -name *.py` or `find . -name *data*.*`
        - `?` specifies one character, such as `find . -name ??_*.*`

- **`grep`** 
    - `grep <pattern> <filename>` searches a file for a **r**egular **e**xpression **p**attern and prints the matching lines
        * The pattern should be in quotation marks to allow for multiple words.
        * The pattern is case-sensitive by default, but you can use the `-i` option to **i**gnore case.
        * You can use wildcards in the filename to search multiple files, but it only searches the working directory (not subdirectories).
    * `grep -r <pattern> <path>` does a **r**ecursive search of the path (checks subdirectories) for matches within files
        * Use `.` for the `<path>` to refer to the working directory.
    * `grep <pattern>` does a **g**lobal search (of your entire computer) for matches
        * Hit `Ctrl + c` if you want to cancel the search.
    * Much more complex string-matching patterns can be used 


#### Input & Output commands
- More are available for you to review using this [reference guide](https://www.cheatography.com/davechild/cheat-sheets/linux-command-line/)
- **`|`** : Pipe: Passes the output of one command into the next
    * `<command 1> | <command 2>` pipes the results from `<command 1>` into `<command 2>`, and then the results of `<command 2>` are printed to the console

- `>` Standard output - writes the output of a command to a destination
    * `<command> > <filename>` takes the output of `<command>` and saves it in `<filename>`
    * This will overwrite the file if it already exists.

- **`>>`** Append: Addes the output of a command to the end of a file  
    * `<command> >> <filename>` takes the output of `<command>` and appends it to `<filename>`
    * This will create the file if it does not yet exist.

## Exercise 4 - Learning more git commands

### Time: 25 minutes

### Purpose

In the below you will learn commands to allow you to create repositories for projects, your capstone and collaborative efforts. These will allow you to share your work with the instructional team for help and support throughout class, keep online versioning and build skills necessary to interact with team


In git there are several basic operations:
- **Init** - This initializes or create a new repository (turn on the ability for file tracking)
- **Add** -  Adds a change in the working directory to the staging area
- **Commit** - Captures a snapshot of the project's currently staged changes. Committed snapshots can be thought of as “safe” versions of a project—Git will never change them unless you explicitly ask it to. 
- **Push** - used to upload local repository content to a remote repository
- **Pull** - Fetch and download content from a remote repository and immediately update the local repository to match that content.
    -**Note** Technically it's a combination of a fetch and a merge command
- **Clone** - Clones a repository into a newly created directory
- **Status** - Displays the state of the working directory and the staging area.


Some advanced operations:
- **Branching**
- **Merging**
- **Rebasing**

To learn more about how these and other git commands work across your local machine (workspace & index) and an online repo check out [my favorite git cheat sheet](http://ndpsoftware.com/git-cheatsheet.html#loc=remote_repo). It includes a flowchart for [recovering from git mistakes as well](http://justinhileman.info/article/git-pretty/git-pretty.png)


#### Building a local git repo

1. In git bash, create a new directory on your Desktop called `myrepo`


2. Now initialize the repository with `git init`. Git will reply: Initialized empty Git repository in <location>.  This is the working directory in which git changes will be tracked.
    
    
3. `ls –A` will show there is a hidden file called .git. This is where all information about the repo is stored.


4. Create a new file called **a.txt** using `touch a.txt`


5. Run `git status`, we should get:
    - On master branch
    - No commits yet
    - Untracked files: with a.txt in red indicating it's a new and untracked file
    
6. To tell Git to take a snapshot of the contents of **a.txt** run `git add a.txt`
    - You can alternatively add all new and modified files at once using `git add .` . Remember the **.** represents the current directory



7. The add command will store the file in a temporary staging area, which Git calls the "index." To confirm the files are staged and ready to be committed, again run `git status` to confirm the file is being tracked. It should appear as green


8. To permanently store the contents of the index in the repository, (i.e., commit these changes to the "HEAD"), you need to run `git commit -m ”Your comment”`
    -**Note** The comment you add should be descriptive and tell you (or your team) why you are making a commit. For now we recommend using the -m argument that lets you add a comment inline. Forgetting it will automatically launch a text editor to allow you to add much more extensive comments. We do not need that for now.
    

9. The command `git log` will allow you to see your comment history



#### Create new git repository
1. On your GitHub account, on the right hand side, select the + button for New repository. 


2. Name your repository myrepo, just like the one on your local computer. (Do not initialize with a README).


3. Notice the instructions to push an existing repository. To send files from our local machine to our remote repository on GitHub, we need to use the command git push. However, you also need to add the name of the remote repo — in this case, we called it origin — and the name of the branch, in this case master.


4. In git bash, while in the local myrepo directory, type `git remote add origin https://git.generalassemb.ly/username/myrepo.git`, replacing the URL with the one shown in your GitHub.


5. Hit Enter and then type `git push –u origin master`.


6. Refresh your GitHub and the files from your computer should appear.


7. Now every time you want to push files, you just need to use git push


### Adding a new file - git push

1. Add a new file called b.txt in your local my repo.


2. Run the command `git status` to see it is there, but untracked


3. Run the command `git add b.txt` to track that file & then run `git status` to confirmed it is being tracked
    -**Note** You can also use `git add .` to add all files to 


4. Once you add and commit this file to git, you can then use git push to add the file to your remote repo.

### Getting an update - git pull

Pull is the opposite of push, meaning you are bringing changes made from a remote repo down to your local repo.

1. In your online github "myrepo", select the button Add A README. You can edit the file and then, near the bottom of the window, use the GUI to commit the file.

2. In git bash run `ls`. Notice this change does not show up in your local copy.  

3. To bring the README.md to your machine do the following.
    - Confirm you are still in the myrepo directory in bash
    - run 'git pull'
    - confirm the file is present using 'ls'
    
### Create a pull request on Github

When working on a project with other collaborators, we all can’t work on same repo at same time, so we use branches. 

With branches, we can create branch off of the main repo, or “master” branch.  There, additions or alterations can be made, and then merge changes back into the master.

Rather than a branch merging automatically, your request goes through an administrator who can review your changes and additions and approve or deny the request.

Even though we are trying to push information into master, it is called a pull request because we are making a request to the administrator to pull our branch into master.


**Instructions**

1. Online in Github create a new repository called test.


2. Initialize the repository with a README.


3. Click "Branch:master" on the left-hand side and create a branch called 'branch-edits' by typing in the box and hitting enter. This should put you on your new branches page.


4. Make sure you're on your new branch and in GitHub create a new '.md' or '.txt' file using the “Create new file” button. Call it what you like and place some text in it.


5. Edit the readme by clicking on the pencil logo on the top right of the readme.  Add a sentence saying “This is a new sentence.”


6. Select “Compare” on the branch-edits branch. Review the changes.


7. Select “Create pull request.” Add a message and select “Create pull request" again.


8. Select ”Merge pull request” and “Confirm merge” to merge the changes into master.  You can ”Delete the branch.” If you were to reject the changes, you can leave a comment and select “Close pull request.”



 

9. The end of today's lesson is to delete your local myrepo then go online and delete your online repo
    - **Local Repo**
        - Either navigate to the the folder in the UI and delete it or complete it from the command line
        - **Note** consider deleting the ".git" folder first, navigate to the folder in bash and notice there is no longer a `master` branch tag anymore
    - **Online Repo**
        - Navigate to the online repo folder
        - Select **Settings** in the upper right
        - Navigate to the bottom of the page and click on **Delete this repository**
        - Enter the repo name then click on **I understand the consequences, delete this repo**
    
10. Let the instructional team know you are done
